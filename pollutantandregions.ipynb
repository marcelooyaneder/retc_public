```python
import pandas as pd
```

```python
df=pd.read_csv('data/df_emisiones_total.csv')  
df
```

```python
df.CONTAMINANTE.unique()
```

```python
pollutant_1='MP10'
pollutant_2= 'Nitrógeno amoniacal (o NH3)'
pollutant_3='Mercurio'
```

```python
ciiu4=pd.read_csv("registros/public_establishments.csv",sep=';')
ciiu4=ciiu4[['ID Establecimiento VU','id_ciiu4','ciiu4','Razón social']]
ciiu4['ID Establecimiento VU'] = ciiu4['ID Establecimiento VU'].str.replace('.', '').astype(int)
ciiu4.columns=['CODIGO.VU','ID_CIIU4','CIIU4','RAZON.SOCIAL']
ciiu4
```

```python
region=pd.read_csv("registros\public_establishments.csv",sep=';')
region=region[['ID Establecimiento VU','comuna','region']]
region['ID Establecimiento VU'] = region['ID Establecimiento VU'].str.replace('.', '').astype(int)
region.columns=['CODIGO.VU','COMUNA','REGION']
region
```

```python
clusterdata=df.query('AÑO > 2015 & `TOTAL.EMISION`> 0' )
clusterdata=pd.merge(clusterdata,ciiu4,on='CODIGO.VU',how='left')
clusterdata=pd.merge(clusterdata,region,on='CODIGO.VU',how='left')
clusterdata
```

```python
clusterdata.query('CONTAMINANTE == @pollutant_3',inplace=True)
#clusterdata.query('REGION == "Metropolitana de Santiago"',inplace=True)
clusterdata
```

```python
# import ClusteringExperiment and init the class
from pycaret.clustering import ClusteringExperiment
exp = ClusteringExperiment()
```

```python
# init setup on exp
exp.setup(clusterdata, session_id = 123, normalize=True,ignore_features=['AÑO','CONTAMINANTE','CIIU4','COMUNA','CODIGO.VU'],categorical_features=['ID_CIIU4','REGION','RAZON.SOCIAL'],use_gpu=True)
```

```python
# train kmeans model
clustermodel = exp.create_model('dbscan')
```

```python
clusterResult = exp.assign_model(clustermodel)
#clusterResult.to_csv('clusterResult.csv',index=False)
```

```python
#exp.save_model(clustermodel, 'output/clustermodel')
```

```python
result=pd.concat([clusterdata,clusterResult[['Cluster']]],axis=1)
result
```
